## Delete columns

```python
X_dropped = X.drop(['CarName','enginelocation','doornumber','fueltype'],axis=1)
X_dropped
```

## Encoding categorical data

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
X_dropped.columns
```

```python
ct = ColumnTransformer(transformers=[('encoder',OneHotEncoder(),[1,2,3,9,10,12])],remainder='passthrough')
X_array = np.array(ct.fit_transform(X_dropped))
print(X_array)
```

## Splitting the dataset into training set and test set

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test,y_train,y_test = train_test_split(X_array,Y,test_size=0.2,random_state=0)
```

## Training the Multiple Linear Regression model on the training set

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor = LinearRegression()
regressor.fit(X_train,np.array(y_train))
```

## Predicting the Test set results

```python
y_pred = regressor.predict(X_test)
```

## Evaluating the model

```python
from sklearn.metrics import mean_absolute_error,r2_score
```

```python
mse = mean_absolute_error(y_test,y_pred)
r2 = r2_score(y_test,y_pred)

print(f"Mean Squared Error (MSE):{mse}")
print(f"R-Squared: {r2}")
```
